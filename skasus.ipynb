Sempurnakan Data Set Universitas Esa Unggul memiliki detail mahasiswa sebagai berikut : 
Ani, Budi, Joni, Jono dan Lono Setiap mahasiswa dating ke kampus mulai dari hari senin sampai 
minggu Untuk melakukan kegiatan kuliah Ani menghabiskan biaya 30000 dikampus setiap datang 
Budi menghabiskan biaya 35000 dikampus setiap datang Joni menghabiskan biaya 20000 
dikampus setiap datang Jono 25000 dan Lono 15000

a) Berapa rata-rata mahasiswa datang pada minggu ini?
b) Kapan biaya tertinggi terjadi? 
c) Hari apa biaya lebih dari 110000?
d) Siapa yang paling banyak datang ke kampus?
e) Siapa yang datang pada hari minggu?
f) Berapa biaya tertinggi dan terendah?
g) Berapa frekuensi datang tertinggi dan terendah?

```python
kedatangan = {
    'Senin': {'Ani': 2},
    'Selasa': {'Budi': 2},
    'Rabu': {'Jono': 4},
    'Kamis': {'Lono': 1},
    'Jumat': {'Joni': 2},
    'Sabtu': {'Ani': 5},
    'Minggu': {'Budi': 2},
}

biaya = {
    'Ani': 30000,
    'Budi': 35000,
    'Joni': 20000,
    'Jono': 25000,
    'Lono': 15000,
}

def ratarata(total_k, biaya, jumlah_hari):
    total = sum(total_k.values())
    jumlah = len(biaya)
    return total / jumlah_hari

def btinggi(total_b):
    hari = max(total_b, key=total_b.get)
    return hari, total_b[hari]

def blebih(total_b):
    return [hari for hari, b in total_b.items() if b > 110000]

def mhs_terbanyak(total_k):
    return max(total_k, key=total_k.get)

def mgg(kedatangan):
    return list(kedatangan['Minggu'].keys())

def btertinggi_terendah(total_b):
    tertinggi = max(total_b.values())
    terendah = min(total_b.values())
    return tertinggi, terendah

def frekuensi(total_k):
    tertinggi = max(total_k.values())
    terendah = min(total_k.values())
    return tertinggi, terendah

total_b = {}
total_k = {mhs: 0 for mhs in biaya}

for hari, hadir in kedatangan.items():
    total_b[hari] = sum(hadir[mhs] * biaya[mhs] for mhs in hadir)
    for mhs, jumlah in hadir.items():
        total_k[mhs] += jumlah

jumlah_hari_dalam_seminggu = len(kedatangan)
rata_rata_minggu = ratarata(total_k, biaya, jumlah_hari_dalam_seminggu)

hari_tinggi, biaya_tinggi = btinggi(total_b)

hari_lebih_110k = blebih(total_b)

mhs_terbanyak = mhs_terbanyak(total_k)

mhs_minggu = mgg(kedatangan)

biaya_tertinggi, biaya_terendah = btertinggi_terendah(total_b)

frekuensi_tertinggi, frekuensi_terendah = frekuensi(total_k)

print(f"Rata-rata kedatangan: {rata_rata_minggu}")
print(f"Biaya tertinggi terjadi pada: {hari_tinggi} dengan biaya: {biaya_tinggi}")
print(f"Hari biaya lebih dari 110000: {hari_lebih_110k}")
print(f"Mahasiswa paling banyak datang: {mhs_terbanyak}")
print(f"Mahasiswa datang pada hari Minggu: {mhs_minggu}")
print(f"Biaya tertinggi: {biaya_tertinggi}, Biaya terendah: {biaya_terendah}")
print(f"Frekuensi datang tertinggi: {frekuensi_tertinggi}, Frekuensi datang terendah: {frekuensi_terendah}")
```
